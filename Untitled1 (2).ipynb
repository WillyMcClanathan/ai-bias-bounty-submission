```python
import pandas as pd

# Load CSVs
train = pd.read_csv("/content/loan_access_dataset.csv")
test = pd.read_csv("/content/test.csv")

# Show dimensions and sample rows
print("âœ… Training shape:", train.shape)
print("âœ… Test shape:", test.shape)
train.head()
```

## Step 2: Preprocessing and Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# Make a copy so we don't mess with original
df = train.copy()

# Convert target column to binary
df["Loan_Approved"] = df["Loan_Approved"].map({"Approved": 1, "Denied": 0})

# Encode all categorical features
encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le

# Prepare data for training
X = df.drop(["ID", "Loan_Approved"], axis=1)
y = df["Loan_Approved"]
X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, random_state=42)

# Train baseline Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_val)
print("âœ… Accuracy:", accuracy_score(y_val, y_pred))
print("\nðŸ“Š Classification Report:\n", classification_report(y_val, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# Make a copy so we don't mess with original
df = train.copy()

# Save original object columns BEFORE mapping
object_cols = df.select_dtypes(include='object').columns.tolist()

# Convert target column to binary
df["Loan_Approved"] = df["Loan_Approved"].map({"Approved": 1, "Denied": 0})

# Remove the now-numeric target from the list
object_cols.remove("Loan_Approved")

# Encode remaining categorical features
encoders = {}
for col in object_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le

# Prepare features and labels
X = df.drop(["ID", "Loan_Approved"], axis=1)
y = df["Loan_Approved"]
X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, random_state=42)

# Train baseline model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_val)
print("âœ… Accuracy:", accuracy_score(y_val, y_pred))
print("\nClassification Report:\n", classification_report(y_val, y_pred))
```

```python
from fairlearn.reductions import ExponentiatedGradient, DemographicParity

# Create the fairness-aware mitigation model
mitigation = ExponentiatedGradient(
    LogisticRegression(max_iter=1000),
    constraints=DemographicParity()
)

# Fit it using training data and the sensitive attribute
mitigation.fit(X_train, y_train, sensitive_features=X_train["Gender"])
```

```python
from fairlearn.metrics import MetricFrame, selection_rate
import matplotlib.pyplot as plt

# Predict with the fairness-mitigated model
y_pred_mitigated = mitigation.predict(X_val)

# Fairness audit for selection rate by gender
metric = MetricFrame(
    metrics=selection_rate,
    y_true=y_val,
    y_pred=y_pred_mitigated,
    sensitive_features=X_val["Gender"]
)

# Plot selection rates
metric.by_group.plot(kind='bar', title='Selection Rate by Gender')
plt.ylabel("Selection Rate")
plt.xlabel("Group")
plt.tight_layout()
plt.savefig("fairness_plot.png")
plt.show()
```

##PHASE 4: Bias Mitigation

```python
from fairlearn.reductions import ExponentiatedGradient, DemographicParity
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Use the original unmitigated data
X_mit = X_train.copy()
y_mit = y_train.copy()
sensitive = train.loc[X_train.index, "Gender"]

# Define fairness constraint and mitigator
constraint = DemographicParity()
mitigator = ExponentiatedGradient(
    LogisticRegression(max_iter=1000), constraints=constraint
)

# Train fairness-aware model
mitigator.fit(X_mit, y_mit, sensitive_features=sensitive)

# Predict on validation set
y_pred_mitigated = mitigator.predict(X_val)

# Evaluate mitigated model
print("âœ… Mitigated Accuracy:", accuracy_score(y_val, y_pred_mitigated))
print("\nClassification Report (Mitigated):\n", classification_report(y_val, y_pred_mitigated))
```

##Fairness Audit of Mitigated Model

```python
# Fairness audit on mitigated predictions
from fairlearn.metrics import MetricFrame, demographic_parity_difference, equalized_odds_difference, selection_rate

# Sensitive attribute
sensitive_gender = train.loc[X_val.index, "Gender"]

# Define metrics
metrics = {
    "accuracy": accuracy_score,
    "recall": recall_score,
    "selection_rate": selection_rate,
}

# Audit the mitigated model
frame_mitigated = MetricFrame(
    metrics=metrics,
    y_true=y_val,
    y_pred=y_pred_mitigated,
    sensitive_features=sensitive_gender
)

# Calculate bias metrics
dp_diff_mitigated = demographic_parity_difference(
    y_val, y_pred_mitigated, sensitive_features=sensitive_gender)
eo_diff_mitigated = equalized_odds_difference(
    y_val, y_pred_mitigated, sensitive_features=sensitive_gender)

# Show results
print("\nâœ… Fairness Audit (Mitigated Model):")
display(frame_mitigated.by_group)
print(f"\nðŸ“Š Demographic Parity Difference: {dp_diff_mitigated:.3f}")
print(f"ðŸ“Š Equalized Odds Difference: {eo_diff_mitigated:.3f}")
```

##Final Predictions on the Test Set

```python
# Make a copy of test set
df_test = test.copy()

# Encode test categorical features using the same encoders
for col, le in encoders.items():
    if col in df_test.columns:
        df_test[col] = le.transform(df_test[col].astype(str))

# Prepare test data (drop non-numeric columns)
X_test = df_test.drop(columns=["ID"])

# âœ… Predict using the fairness-mitigated model
y_test_pred = mitigator.predict(X_test)

# Prepare submission file
submission = pd.DataFrame({
    "ID": df_test["ID"],
    "Loan_Approved": y_test_pred
})

# Map numeric predictions back to Yes/No
submission["Loan_Approved"] = submission["Loan_Approved"].map({1: "Approved", 0: "Denied"})

# Export to CSV
submission.to_csv("submission.csv", index=False)
print("âœ… submission.csv created successfully!")
```
